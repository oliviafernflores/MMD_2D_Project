```python
import dadi
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(1, figsize = (10,6))
fig.clear()

ax = fig.add_subplot(2, 3, 1)
ax.set_title('New Synonymous SFS')
dadi.Plotting.plot_single_2d_sfs(dadi.Spectrum.from_file('IRA_FRA_syn_unfolded.fs'), ax = ax)

ax = fig.add_subplot(2, 3, 2)
ax.set_title('Old Synonymous SFS')
dadi.Plotting.plot_single_2d_sfs(dadi.Spectrum.from_file('/Users/oliviafernflores/Documents/2D_demographics_DFE/MMD_2D_Project/FRA_IRA/IRA_FRA_syn_unfolded.fs'), ax = ax)

residuals = (dadi.Spectrum.from_file('/Users/oliviafernflores/Documents/2D_demographics_DFE/MMD_2D_Project/FRA_IRA/IRA_FRA_syn_unfolded.fs')) - (dadi.Spectrum.from_file('IRA_FRA_syn_unfolded.fs'))

ax = fig.add_subplot(2, 3, 3)
ax.set_title('Difference in SFS (old - new)')
dadi.Plotting.plot_2d_resid(residuals, resid_range = 3, ax = ax)

fig.tight_layout()

plt.savefig('IRA_FRA_syn_fs_comparison.png')
```
