```python
import dadi
```

```python
import matplotlib.pyplot as plt
```

```python
import pickle
```

```python
dd_syn = pickle.load(open('mus_musculus_domesticus.dd_syn_correct.bpkl', 'rb'))
```

```python
dd_syn
```

```python
fs_FRA_IRA = dadi.Spectrum.from_data_dict(dd_syn, ['Mmd_FRA', 'Mmd_IRA'], [16, 10])
```

```python
fig = plt.figure(1, figsize = (10,6))
fig.clear()

ax = fig.add_subplot(2, 3, 1)
ax.set_title('Synonymous SFS From Syn Data Dictionary')
dadi.Plotting.plot_single_2d_sfs(fs_FRA_IRA, ax = ax, vmin=1)

plt.savefig('IRA_FRA_syn_from_dd_fs.png')
```

```python
fs_FRA_IRA
```

```python
fs_FRA = dadi.Spectrum.from_data_dict(dd_syn, ['Mmd_FRA'], [16])
```

```python
fs_FRA
```

```python
fig = plt.figure(1, figsize = (10,6))
fig.clear()

ax = fig.add_subplot(2, 3, 1)
ax.set_title('FRA SFS from Synonymous Data Dictionary')
dadi.Plotting.plot_1d_fs(fs_FRA)
plt.savefig('FRA_syn_from_dd_fs.png')
```
